```python
#translating from RNA to protein
def translate_rna(rna_string):
    codon_table = {
        "UUU": "F", "CUU": "L", "AUU": "I", "GUU": "V",
        "UUC": "F", "CUC": "L", "AUC": "I", "GUC": "V",
        "UUA": "L", "CUA": "L", "AUA": "I", "GUA": "V",
        "UUG": "L", "CUG": "L", "AUG": "M", "GUG": "V",
        "UCU": "S", "CCU": "P", "ACU": "T", "GCU": "A",
        "UCC": "S", "CCC": "P", "ACC": "T", "GCC": "A",
        "UCA": "S", "CCA": "P", "ACA": "T", "GCA": "A",
        "UCG": "S", "CCG": "P", "ACG": "T", "GCG": "A",
        "UAU": "Y", "CAU": "H", "AAU": "N", "GAU": "D",
        "UAC": "Y", "CAC": "H", "AAC": "N", "GAC": "D",
        "UAA": "*", "CAA": "Q", "AAA": "K", "GAA": "E",
        "UAG": "*", "CAG": "Q", "AAG": "K", "GAG": "E",
        "UGU": "C", "CGU": "R", "AGU": "S", "GGU": "G",
        "UGC": "C", "CGC": "R", "AGC": "S", "GGC": "G",
        "UGA": "*", "CGA": "R", "AGA": "R", "GGA": "G",
        "UGG": "W", "CGG": "R", "AGG": "R", "GGG": "G"
    }

    protein_string = ""
    for i in range(0, len(rna_string), 3):
        codon = rna_string[i:i+3]
        if codon in codon_table:
            amino_acid = codon_table[codon]
            if amino_acid == "*":
                break  # Stop codon reached, end translation
            protein_string += amino_acid

    return protein_string

translate_rna("AUGUGUGCCAUCUUUGGGGAGUCUUGGGUUGCACGGUGGCUCUACUUGUCAACCACGAUGCUAACCAAUGGUUACAUAAAGUAUGUGUACCAUGAGCGGUAUACGAGAGCGUGUGUUUCAAAUGUAAGUCCCUUAGCUCCACUAGGGGACUCGAACAAGAAGAGUAUGGUCACAAAGCGUGGGACAGGUAGGUCACAGCCUAUACACCGCUUAAUAAACUUGGAGCUGAAGUCGCCCUCUGGAAGCGAGCUAAUUAAUAUAGAUGAGUCACGACUUCCUUUACUCCUUCCACAAGGGAACGCGACGCCAGGUACUGCGAUACGAAUUAUCACCUCUCGCAUAAAACAAGCGUGUCGGAGUAUUAGACCCAACUCUAGCAUUCGUCCGGCUCCAAAUAACGCUGCUGUCCGUCAGAGGCAUUCUGGGUCUUGGAAUGAGUGGUAUUUGUGCGCACAUCAAAUUCCGUGUUGGUUAGUGAAAAGUCGAUCGCAGGUCUGGAUUCCGAUCCACGAUACACUUCUUUUACUCAGCGGCUUGGAACGAUGUCCCUUCCAUACGGACAGUCUUAAGAGUAUAUUAAGACAAGAAGAGGGUUCCUAUCGGCAGACUAUUUCCCAUCCCUAUGAAGGACAGAGAAUACUAAAAACCUUCUUCGCUUUGAGGCGUGAUGCCCGCUUCAUAAUCCGAGCUAUGAUCGAGACAGCUAGGACAACCCGCCGGUCGUUCGUAAUUACAGGUUUUAUUGGUAGAUCGGGGACAGCCAGCGGAAAUCGAGUCGUUCGAUUUGACGUGAAAGAGAAUGUUCACUUUACUGUUGAAACACGCUGGUUCAAAGUUACUGGGGCACAGCGGCAGAAGCGUAGUUCGCCCACCCCCUGGACAUUGUUGACGCCGACGAAAGUUAGAACAGUUCCCGCUGCUCUCAAGAGUGCAUUACUUGCUGCGUUAUCGCGGAGGGCGAUGACCUUGCGCAACUCUCUCGAUUCCCGACAUCAUUCACUUCUUUAUUCUCGUGACGUCCGCUACAACUCGAAGGCAUCCCACUCUCGCGUGAUAGGCACUUUUGUAUGUUCGCUCAUUCGUACUCUGGUUGGACCCCGCAUGUAUUCCCCACGCUACUGGGACUCAGGUAUUCGACGAUGUGGUCCCUGUAGCCACCAAAUGUUGAUCUUAACCGUAUUGAUCAGGACAUACUGCAUUUGUGAAACCCCCUGUAAAUCCAAAGGCAGUACCGACCGCAUGAUGAACUUGUUCUUCCCGAAUCCCAAGAUCCGUCAACUAGUCGAUCACGUCUCCUCGCCCCAGCUAUUGGGCAGGGUUGGGGACAUCCGAAUUUUAAAAAGCGAGUCCCUAUAUGUGGCACGCAUUGUGGUCUUUAAAGAUACUGGAUGGAAACGCGUAUCCCGUACAUUCGGGUCUAUCCGUUGCGACCCCAAAAUGUUCGUAGCGUAUCUGAUCUUCCGACGCUCAAACCACGCUACGAAAGGUAAACCGCCAUUUUCGGCCGUGCGACCAUCGUGCGUGCUAGGCGAAUUCUUAGCUACUAAACGAGGGGGGCCCUACUCGAUAUCGCUACUGUCUGAAGGGGGCCGGACGCCUCAAAAGGUACGCGUUUACGGCGCGGGCAGGGGCCGUAGCGACGUGGGAACCGAAAGGGCUUGGUCUCACGAUGAGAUCAAGUGCCGUCGAAAUCUGGCCCACGGUCUAGCCCUUGAGAGAGGGUAUCUGCUAUUGCUGGGUGCUGGUUCUGGGAAACCCGCACUAAUACUUAUUGCCCCGAGUAUUAGGAGCACCUCCUUUCUCACAUUCAUGUCUACUAACUUGGUUGAGCUGCAGUCUUCCUACAACGCUCGCAUUGCUCCAAUGUCCGAGUUCCGUCCAGACCUCUAUAAGCGUAGUUAUGUACUAGCCAUGGCAUCACGGCUCUGGAAUUCUGGGUGUAACUGCUCAUGGCGCAGAGCCCUGUGUAUUGAUUCCCUAUUGAUACUUCUAGGUCCGGCUGGCAUCCACUUACGCAUAACGGGGCUCGUGCUGUUAACCACCUGGUCCAGCAGAAUCUUGCUAAAGUCUGGUCUGGGAACGAGUCGCCGACUGUAUGUAGCUCAGACACAUUCCUUAUGCAAUCCUAAAAAAACUAUCCUUGGUUUAGGGAUAAUUUACUUAGAAACAGCGUGUUGCGUCCGAGAUUAUGGCUUACAUGUUUUCUCAGCGGACCUAGAAGAGAUCCUUCGCAGCCAAUUCAUGAGACGAAAUGUCCAACCGGGGGAACAAUUGGCUACCUUCCAUCUUAUCGACCGAGUCUCUGAACUGGGGGCUUUGUGGGAAAAUCGGGAUGGUGCUGUCAAUGUAGGCGGAUUGAGAUCUAGCUGGCCCGGGUUAGGACGCGUGUCGGACCUCAUUAGCUUUGUGGAACGUGUGUUCGUCGGCUGGUGUGGACCGGGAUCUCCUCAAUACGCGGUUCACCCAGUGGCGUCAAUAGGCUUUAUGACUCUGUGGAUGCCGCGGACGAACCCCCUUCACAUGAGGACAACCUUACAGUCAUGCUAUCUUGUUAAACAGUUUCUGCGCUGGGGGGGCAACACUUCAACGCACCCGAAUUUCUCUGUUUGCCUCUGCUCUCCGCCCUGUAUCUCAAUAACUCUCCUGAGACUUGCGCGCCAGGACCCUUGGUCAGGAGAUGUUACAUCUUGUACAGCAGUCGUUCGUACCGACAGAACCAGCGAUUCGACAUCCCCACGGGCGUCUGGUAACGCACGCUUGUGUUUACAAUCGGACAUGAAGCAGGUAGCUGCAGAGUUGGCCCCCUGUCUUUCCACAGUGAUAGGACCGAGUCCGCUGGAAUUGAGGGAAACCCCAUCGUGUGCCCGAAACCUAAAUCGAUAUGUAUCUUGGUCGGUUGUAGGGGGGAGCCGUGCUGCAUUAACACAUGUCCCCAAACGAGAUGGUAGACCCCCUGAUCCAGGCACCCCGAAGUGUAAUAAAGGGUCUAUACAUUCUUUUCACUUACAUAGUGUACAUAGACUUAUUCUAGCAUAUUGCGUCUCGCUGUGGAGCGCCCUGUGCGGUACACUUACCGCCCAAGGGCCAUUGGCUCCUCCACCCAACUAUUUUGAGCUUCGCAACUAUCGCGCAUGCAGUUACCACGGCAGCCUAGUGUUCACGGUUCAUGUAGAUUAUGUGAGUUAUGCAUCACUUUUCUAUUUAACGUCAUCUAUGGCGAAAUGUUGGAGGCUGCUCAAACUCCAUUUGCCUGCGCAGAGCAAAAUAUCUAUACCGAGCGGAGCUGCCCCAAACAGUGCAGCCACUCGUGGAUUCCGGCACUGGUUUAUCUGUGUCAUACUUAUAUUUAGGCGGAAUAAGUCAGCUCACGGAUGGAGGUACUUAGUACUGAUUCAUCUGACUCGGACAUUUCGCCUAGAUGCACGAUCUAUUCCAAGGGUUACCGUGUUAGGAUGCCAGUCCACCCGUCAUUCGAGUACAGAAUGGGCUCUCUUUGGCAAAUGGAUUGCCGGGGGAGUGGUACCCUUGCGCAGUUCUGCUACCGUUAUAUUACGGGAGAUGAAGGGGUACGAACACUGGUCGCGUUGCUUGAGGCCUGUACCUACUGCCCUAUAUCAGGCUUGUAUCGGUCAUUCCUGUCAGCGUGUCCUGAAAUUCCUGGGAUUACUUUCCAUGAUAUUGGGUGAUGCUUCAGCACCUAGCGGUGUUCUCGUAACGCUUCAAUACCAGUGUAGGAAUAAGAGCGUGGGCUUUUCUAAGGGUCCCCGCGCCAUUCGCUGCUCUCCGCACGCCAACAGGCAGCAGCACUUUAGAGCUUCCGAGCACGUUAGGUUUUCAUUUGACUCUUUACACUACGCGCCUGGUGACGUUUUGGAAACUUACUACUCCUGUCAUAAUAUUCAUUGCAACAAAAUCCCACUUGGCCAUCCGGCCGCUAAGUAUCGUUGGGUUGAUUUUAAGGUUGGGAACUCUAGGCUCCAGAAGAGGCAACUACUUCCCCCGCUCAAGUCGCGCGUUGGUGAUAGCCAAAGCGUGAAAUGGUCCUUGCGCGCCCGCACCAAGGCCAGGCUGAAACCCAUGCGGAUACGCAUGAAACAAGCGGAGAGCGCAUGCUCUAUUGCUAGAACGUUGGCCUACGCAGGAACUAUCUCACUUCAGCCGGCUGUGCGCUUUGCGCUAAGUUCGAGGAAGGACGCCUGGUACCCUCGCGCAUUACUUAUGCAAACUAGGCGUCCAUUCCAAAUAUGCUGCGGAGCCUUAACGGUUCCUCAGAGGUGUAUGGGAGGAAAAAAAGGGGUGUGUCAGGAUGACUGCCCUUCAGUUUCUCACAGCUGCGAUUGCGUACGCAUACAUGUGACCGCGCGUUCUUUCUGCUCAAAAUCGUCUUUCCGCCCCAUUGAAGCUAAAGUCGUCCUGUACCACCGCCAGCCGCCGUAUUACCCCAUGAUAUCCAUUUAUUCCCGCCUUUUAAGGAUUACAGACUCCCUGGCCCUGCGCAGCCGGUGGCUAACUUUUCUCAGUGCUAUUAAGGUGUUUCUCUCGCGCGUCGCUUGUUCAUAUAGCUUAUACUCCCGUUGUGCACAUCCCCCGCCUGCGCCAUCAGCCCAAUGUAUGCUGCGGGGGGCUUAUGGAAUUUGUGGUAGGGCUAUCUGGAAGGUGAUAUACCGGUUCGAUUUACCUUCACCCAAAACCAUGAUUAUUACAACACUGUCCAGAGCCGGGUUUCGUGAUCUAAUUAGAGCAUUAUGUAAGAGGCACGCGACGGACUCAUCAGGCUAUAAUGCUCGUAUAUCAAGAUCAACUGCGGCGGUUUCAAGCUCAAGGCUUCUGCGCCUCUCUGCCAUCCCACAUUCUGUGUUAUUUUCCGCAAGAAGCCAAGAACUUUCGUCUCUGCUACGGAACGACUUUUCGUAUCAGAUCUCAGGCUCCAUACCCCAGUGUGGCCAUGCUACAUCGUUGUGCGCCAUACCCGCUUUGAGGGUUAUCAGUCUAUGGACCACUUAUGGAGGUCCAUCCUUUCAUUGUAUUGUCAGCGCUAGGUUUAGAGUCCCAUGCCCCGUGGUCUCAGCAGGCGAGUGGCAUCAGAAGUGGCACCUCAUGAUUACGGGUGUAGCCUCGAGUCGGAGGUUUUCACAAGGGUCUGGGGGUAGGCUCUGGAUUAUACACUAUAAGGUCAGGACACGUUUCGGAAUUAGAUGUUCAAUUCCCGUACGUCUAGAUAGUCUUCAUGUCUUCAUUCGGCUAGGUGUCUUUCUUAAGUUGUUUCAGGUCAGCGUACAGGGCACCACUCAAAUCCAACUAGGAUUCGGAGAGUUUUACAUUUCUGACAUAAAGCGUUCAGAUGCAAUAUCACUUAUUUAUAGGGUUGGCUUUCUCGCCUCACGGAAGGUCAGUUUAUUUGAAUGGUACUUAACUUGCGCCCGCAAACUAGCCCUAAAUUCUCCUUUAAUAAUCAUCCUAUCCUCCGAUAGAAAGCUUUGCAGGUACUACUUCAUCGGUACCCGCCGAGGUUACGGACCCCGAUGCAGUUUGAGGGGUAAUAAGAUUUCGCUAAAGUCUCAUGGUCCCUCCGCACAUCUACAGGUACGAUUCGGUCGGGGGACGACACUAGUUAUCACGUCCGAUUGUAAAGCUAUAGCGUCUUUUAGGCCUAUCGGCCUACACUGCGUGGCUCAUAACCGCAAUAGGCGCGGCGGAGUCUGGCUGGGUAGAUUCAGACGUGACCGGACAGUAACGGUUACGCACCCAAUUAUCUUUUUGGAGUCCGUACCAGAGGGCACUGAGCGAGAAAACCAAGGGGACCUUGUCAGCAUGCCCCGCAACUCUUACAUGUAUAAGUGGCAAGCAUAUUUUAGGGAGCCAGUGGAGGCUCGAAAUGUGCGGGUACUCCGCUUAGGCCUUUUUGAGCGAGUCACCAAUGACAUUCAAAUUUUGCUUACUCACGGUGUGAUCACCCAGAAGAAAGAGUGGACAGUAGAUCCGCACUAUCGGAAUCCUAUAGGGCUAUCGCCUUUAUAUACUAAGCGACUCGUUGUACAGGCUUCAACACUAUUUUGCUCACUAGCACUUCUCCCCAUACACGCUUCAGCUAUUCUAACACUUGUCAGUACAACGGACAUACGUGUUUCAGCUUGGCAACCUCCGGUACUCAUCAGAAAAUUUCAUACCCUCUAUAAAAGCACCACUUGUAGGAGAAAACCUAACGGCUGUCUUGACAAAUCAGUUAUUGACUGUAGAGUCUUGACAACUCAGGAAAUGGGCGCACAUCUUAAAUAUGUAGGUCGGGGGCUAAGACUAAUAGAAUGGUGUAGGCUCUUGGGGUUGCCAUGUCAACUAAGCUUGUGGUCUACUGAUCCUUAUCAAUUUAAGGAACCUGAGAGAAAUGUUCCUCUAAUCUAUUUAGACCUGGGAUCCGGGUCCCACCCACCAAACGCGGUUCGUAGACGCCAUCCCACACAAUGUAUGAAUGGCGUAUUAAGCGUCGUAGGUAGGCCUCGCGCGCACCCAUCUGAGAGACGGACCGGGUCUUAUCGUAAACGGGUUCUAAAUUUGGCUGGUCGACUGUCUUUAGAUGUAGCACUUUAUCACUCAGACGUUAAUCUCCUCGCCAUGCUAUUCGAUUCAGCUCGCGGUCUUCUGCCUAUGCACAAGCCGUUCCGAUCGUAUUGCGUAUUUAUUUAUCUGUUUUACUCGGAAAACUGCACGCUAUUCCCGGUAUGGGCUGUGAUCAGAGUUCAGUCUGAUUUGCCAGGGCAUGUCCUGACGACAGCGGUACUAAGAGAUGUUAUAGCCACCAGGUGCGCGUGUAUGCUUACCAUCAAGUACGCCUUGCGGUUUGGUAUAUUAUGGAGCACCAUGGGGGCCGAUAUGCAGGGAGUCUCUAAACAACGUCAUGUGACCUGUGCUUUGGUCGGUGUUAGACAUUGCUCACGGUCAGUAACCGUCGCGAAUAAUCAGUGCGUGCCCGACAUGAUUUAUAGACGUAGCGCUCUCUGCCUUUGUAUCUUACUACUUUUACAACCUACGCAGCCUGCUAGGAGAGCUUGCCUGUGUCGAACAAGCUUAAAGCUCCUGUGCGCUCUAACGGUUAGGCUGGGGACCUUGGCCCGGCUCUCCCAGUAUCUAUUUAACGGUUUGGGAAAUGAUGAGAAAACUGCAUGUCUCCCCGAGAGAAUUUUCUUGUCAGAUAACUACGUAACCUACUCAUUUCCUGUAUAUAUCACCUCGGUUAGCAAUUGUGUAAUAACACCUGAUCAAUGCAUGCUAGAUGCCGGAACGGCGGGCGGAUUGUACGAUGCUUUCCCUAUCUCAGUGCGGCUGAAUGUAAUGAGACCGAGAAGUCAUUACCCCACGAUUUUAGGAGACCAUACGUCACGUUACGGACGAGAACAUGGCCUGGCGCAGUUUCAGACCUCUCAUUGUGCCCCCAUUAUCGGCGGUUGGAGCAGGGGAGGGUCACCUAACCGUAUGGUAUGGGGGGUUAAAAAUCCUGGGGAGGCGAUCUUAGCGAAUUUGGUCAUGAACAGAUGUCUCGGUGCAGACGAAUAUUAUUCUACCCGUAUACAUCACGGACUACAAACACGCGAGUCACAAUACCUGGAGAGACUCCAAUUCCCAGGCAAUACCCUCGACGCCUGCGCAAUAUCGACUGCAGUAUUCCUUAGGACGCACUGCCGAAACCAGGGUGCAACGUCGAUUGGUAAAUUCGAAGGCCAAGUCGCAAGUCUGCUGCAAUUGCUCCCCUUCUGCGGUGCGCUACAUAUCAGAACAGUCCUGCGCCGAAUGAUAAAGGUGCGCGAACCGCAGCGCCCAAAGCGUUUGGCAGGAACUACGCCGUACAACACCUGGUUCUAUAGACCGCCUGGAUAUACCAAAGAGAAAAUGGUCCGAGGAGCAAUAAUCUGUUCUGGGGACGGUGCAGUGAGGUCGAGUAUGCGACCAACUUCGAGAGAAGUGCAGGUCGGUCAGAUUGCGAAGAGGGGGCCUUGUGAUGAGAGCUGA")
```

```python
def fib(n, m, M):
    assert M >= 3
    if m == 1:
        if n == 1:
            return(1)
        elif n == 2:
            return(0)
        elif n >= 3:
            fn1 = 0
            for i in range(2,M+1):
                fn1 += fib(n-1, i, M)
            return(fn1)
    if n < m :
        return(0)
    else:
        return(fib(n-1, m-1, M))

def FIB(N, M):
    fn = 0
    for i in range(1,M+1):
        fn += fib(N, i, M)
    return(fn)

# Solution 2: use a more simple method
# see https://math.stackexchange.com/questions/511565/fibonacci-with-mortal-bunnies
# and https://duphan.wordpress.com/2015/07/10/dynamic-programming-and-mortal-fibonacci-rabbits/
def fib(passed_month:int, max_age:int, mature_month:int = 2):
    current_generation = [1] + [0] * (max_age-1)
    for i in range(passed_month-1):
        current_generation = [sum(current_generation[mature_month-1:])] + current_generation[:-1]
    return(current_generation)

print(fib(6,3))
print(sum(fib(6,3)))
```
